```python
h, w = 300, 300
image = np.zeros((h, w, 3), np.uint8)

row = (h - 80) // 2
col = (w - 80) // 2

for i in range (80):
  for j in range (80):
    image[row + i, col + j] = (0, 0, 255)

row2 = (h - 60) // 2
col2 = (w - 60) // 2
for i in range (60):
  for j in range (60):
    image[row2 + i, col2 + j] = (0, 255, 0)

row3 = (h - 40) // 2
col3 = (w - 40) // 2
for i in range (40):
  for j in range (40):
    image[row3 + i, col3 + j] = (255, 0, 0)

row4 = (h - 20) // 2
col4 = (w - 20) // 2
for i in range (20):
  for j in range (20):
    image[row4 + i, col4 + j] = (0, 0, 0)

cv2_imshow(image)
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Función Seno
def generate_sine_wave(amplitude, frequency, phase, num_samples):
    t = np.linspace(0, 2*np.pi, num_samples)
    y = amplitude * np.sin(2 * np.pi * frequency * t + phase)
    return t, y

# Parámetros de la función seno
amplitude = 50
frequency = 1
phase = 0
num_samples = 500

# Generar la función seno
t, y = generate_sine_wave(amplitude, frequency, phase, num_samples)

# Crear una imagen para visualizar la función seno
image_size = (500, 500, 3)
image = np.zeros(image_size, dtype=np.uint8)

# Escalar y desplazar la función para que se ajuste a la imagen
y_scaled = y * (image_size[1] // 2) / amplitude
y_shifted = y_scaled + (image_size[0] // 2)

# Dibujar la función seno en la imagen
for i in range(1, num_samples):
    cv2.line(image, (int(t[i-1] * image_size[1] / (2 * np.pi)), int(y_shifted[i-1])),
             (int(t[i] * image_size[1] / (2 * np.pi)), int(y_shifted[i])), (255, 255, 255), 2)

# Mostrar la imagen con Matplotlib (compatible con Colab)
plt.imshow(image)
plt.axis('off')  # Desactiva los ejes
plt.show()

```
